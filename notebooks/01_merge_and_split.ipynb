#### __1.3 Merge Tables__

```python
# remove count_x columns
demographic = demographic.drop(columns=['Count'])
location = location.drop(columns=['Count'])
services = services.drop(columns=['Count'])
status = status.drop(columns=['Count'])
```

```python
# Merge tables Demographic, Location, Services and Status on "Customer ID"

df = demographic.merge(location, on='Customer ID', how='inner') \
                       .merge(services, on='Customer ID', how='inner') \
                       .merge(status, on='Customer ID', how='inner')
```

```python
print(df.shape)
df.head(2)
```

```python
# Merge df with Population table
df = df.merge(population, on='Zip Code', how='left')
```

```python
# Set "Customer ID" as an index
df.set_index('Customer ID', inplace = True)
```

## 1.4. Basic Exploration

```python
# Check the dataframe info (datatypes, non-null values)
df.info()
```

```python
# Check for duplicates
if df.duplicated().any():
    display(df[df.duplicated(keep=False)])
else:
    print('No duplicates found')
```

```python
#Check the name of the columns
print(df.columns)
```

```python
# Check for columns with identical content
dup_content_mask = df.T.duplicated(keep=False)
dupe_content_groups = (
    df.T.assign(_g=df.T.apply(lambda s: pd.util.hash_pandas_object(s, index=False).sum(), axis=1))
      .groupby("_g").apply(lambda g: list(g.index)).tolist()
)

dupe_content_groups = [g for g in dupe_content_groups if len(g) > 1]
print('Identical-content groups:', dupe_content_groups)
```

```python
# Drop repeated column "Quarter_x" and rename "Quarter_y" to "Quarter"
df = df.drop(columns=['Quarter_x'])
df = df.rename(columns={'Quarter_y': 'Quarter'})
```
